```python
import pandas as pd
import json
```

```python
df = pd.read_json('accounting jobs.json')
df.head(1)
```

```python
allJobs = []
jobTypes = ['Full-time', 'Part-time', 'Contract', 'Temporary', 'Commission', 'Internship']
permissionTypes = ['U.S. Citizen']
```

```python
def cleanJson (fileName, toPend, major, degree):
    with open(fileName) as file:
        data = json.load(file)
    for job in data:
        fullDescription = {}
        fullDescription['position'] = job['title']
        description = set(job['description'].lower().split())
        for keyWord in jobTypes:
            if keyWord.lower() in description:
                fullDescription['type'] = keyWord
        if 'type' not in fullDescription:
            fullDescription['type'] = 'Full-time'
        fullDescription['description'] = job['description'].replace('\n', '<br>')
        fullDescription['summary'] = job['description'][:1500] + '...'
        fullDescription['company'] = job['company']
        fullDescription['url'] = job['job url']
        fullDescription['location'] = job['location']
        if 'citizens' in description or 'citizen' in description or 'lpr' in description or 'residents' in description or 'resident' in description:
            fullDescription['authorization'] = 'U.S. Citizens'
        else:
            fullDescription['authorization'] = 'Other'
        fullDescription['major'] = major
        fullDescription['degree'] = degree
        fullDescription['paid'] = True
        toPend.append(fullDescription)
```

```python
allJobs = []
cleanJson('accounting jobs.json', allJobs, ["MATH", "ECONOMICS", "STATISTICS"], "MS")
cleanJson('banker jobs.json', allJobs, ["MATH", "ECONOMICS", "STATISTICS"], "MS")
cleanJson('construction jobs.json', allJobs, ["ARCHITECTURE"], "NOT REQUIRED")
cleanJson('data jobs.json', allJobs, ["COMPUTER SCIENCE", "STATISTICS"], "PHD")
cleanJson('driver jobs.json', allJobs, ["NOT LISTED"], "NOT REQUIRED")
cleanJson('economist jobs.json', allJobs, ["MATH", "ECONOMICS", "STATISTICS"], "PHD")
cleanJson('government jobs.json', allJobs, ["ANTHROPOLOGY", "CIVIL AND ENVIRONMENTAL ENGINEERING", "PHILOSOPHY", "COMMUNICATION"], "MBA")
cleanJson('hardware jobs.json', allJobs, ["ELECTRIC ENGINEERING", "COMPUTER SCIENCE", "COMPUTER ENGINEERING"], "BS")
cleanJson('healthcare jobs.json', allJobs, ["BIOLOGY", "BIOENGINEERING", "CHEMISTRY"], "PHD")
cleanJson('HR jobs.json', allJobs, ["COMMUNICATION", "BUSINESS"], "MBA")
cleanJson('libarian jobs.json', allJobs, ["CHINESE", "ENGLISH", "HISTORY", "PHILOSOPHY"], "BA")
cleanJson('marketing jobs.json', allJobs, ["CIVIL AND ENVIRONMENTAL ENGINEERING", "COMMUNICATION"], "BA")
cleanJson('pharmacist jobs.json', allJobs, ["BIOLOGY", "BIOENGINEERING", "CHEMISTRY"], "PHD")
cleanJson('security jobs.json', allJobs, ["NOT LISTED"], "NOT REQUIRED")
cleanJson('software jobs.json', allJobs, ["COMPUTER SCIENCE", "COMPUTER ENGINEERING"], "BS")
cleanJson('tutoring jobs.json', allJobs, ["COMMUNICATION"], "NOT REQUIRED")

len(allJobs)
```

```python
with open ('Jobs Data.json', 'w') as file:
    json.dump(allJobs, file)
```
